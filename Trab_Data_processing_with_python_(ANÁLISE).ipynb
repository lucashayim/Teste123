```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dicionário de dados do arquivo de PowerPoint fornecido
# com os nomes das colunas na ordem correta [cite: 4, 7, 10, 13, 16, 19, 22, 24, 27, 29, 32, 34, 36, 39, 42, 45, 48, 51, 54, 57, 60, 63, 66, 69, 72, 75, 77, 80, 83, 86, 89, 92, 94, 96, 99, 102, 105, 108, 111, 114, 117, 120, 123, 126, 129, 132, 135, 138, 141, 144, 147, 150, 153, 156, 159, 162, 163]
colunas = [
    'ID_CLIENTE', 'TIPO_FUNCIONARIO', 'DIA_PAGAMENTO', 'TIPO_ENVIO_APLICACAO',
    'QUANT_CARTOES_ADICIONAIS', 'TIPO_ENDERECO_POSTAL', 'SEXO', 'ESTADO_CIVIL',
    'QUANT_DEPENDENTES', 'NIVEL_EDUCACIONAL', 'ESTADO_NASCIMENTO', 'CIDADE_NASCIMENTO',
    'NACIONALIDADE', 'ESTADO_RESIDENCIAL', 'CIDADE_RESIDENCIAL', 'BAIRRO_RESIDENCIAL',
    'FLAG_TELEFONE_RESIDENCIAL', 'CODIGO_AREA_TELEFONE_RESIDENCIAL', 'TIPO_RESIDENCIA',
    'MESES_RESIDENCIA', 'FLAG_TELEFONE_MOVEL', 'FLAG_EMAIL', 'RENDA_PESSOAL_MENSAL',
    'OUTRAS_RENDAS', 'FLAG_VISA', 'FLAG_MASTERCARD', 'FLAG_DINERS',
    'FLAG_AMERICAN_EXPRESS', 'FLAG_OUTROS_CARTOES', 'QUANT_CONTAS_BANCARIAS',
    'QUANT_CONTAS_BANCARIAS_ESPECIAIS', 'VALOR_PATRIMONIO_PESSOAL', 'QUANT_CARROS',
    'EMPRESA', 'ESTADO_PROFISSIONAL', 'CIDADE_PROFISSIONAL', 'BAIRRO_PROFISSIONAL',
    'FLAG_TELEFONE_PROFISSIONAL', 'CODIGO_AREA_TELEFONE_PROFISSIONAL',
    'MESES_NO_TRABALHO', 'CODIGO_PROFISSAO', 'TIPO_OCUPACAO',
    'CODIGO_PROFISSAO_CONJUGE', 'NIVEL_EDUCACIONAL_CONJUGE',
    'FLAG_DOCUMENTO_RESIDENCIAL', 'FLAG_RG', 'FLAG_CPF', 'FLAG_COMPROVANTE_RENDA',
    'PRODUTO', 'FLAG_REGISTRO_ACSP', 'IDADE', 'CEP_RESIDENCIAL_3',
    'CEP_PROFISSIONAL_3', 'ROTULO_ALVO_MAU=1'
]
```

```python
# Carregando o arquivo sem cabeçalho e atribuindo os nomes das colunas
df = pd.read_csv(
    "https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/credit.csv",
    encoding='unicode_escape',
    sep='\t',
    header=None,
    names=colunas
)
```

```python
# Renomeando a coluna alvo para facilitar a análise
df = df.rename(columns={'ROTULO_ALVO_MAU=1': 'classificacao'})
```

```python
# Exibindo as primeiras linhas do DataFrame com os novos nomes de colunas
print("Primeiras linhas do DataFrame:")
print(df.head())
```

```python
# Exibindo as informações gerais do DataFrame para verificar tipos de dados e valores nulos
print("\nInformações do DataFrame:")
print(df.info())
```

```python
# Análise da distribuição da variável alvo
print("\nDistribuição da variável de Classificação:")
print(df['classificacao'].value_counts())
```

```python
# Plotando a distribuição da variável alvo
plt.figure(figsize=(8, 6))
sns.countplot(x='classificacao', data=df)
plt.title('Distribuição de Bons (0) e Maus (1) Pagadores')
plt.xlabel('Classificação')
plt.ylabel('Contagem')
plt.xticks(ticks=[0, 1], labels=['Bons Pagadores', 'Maus Pagadores'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('distribuicao_classificacao.png')
plt.show()
```

```python
##################### ABAIXO COMEÇAM AS RESPOSTAS A SEREM ANALISADAS DA SEGUNDA PARTE #####################
```

```python
# Tabela de contagem cruzada para ver a distribuição (( DADOS A SERAM ANALISADOS PARA RESPONDER A PRIMEIRA QUESTÃO DA SEGUNDA PARTE!!!!))



print("Tabela de Frequência - Sexo x Classificação:")
print(pd.crosstab(df['SEXO'], df['classificacao']))
```

```python
# Tabela de contagem cruzada normalizada para ver as proporções
print("\nTabela de Proporção (%) - Sexo x Classificação:")
print(pd.crosstab(df['SEXO'], df['classificacao'], normalize='index') * 100)
```

```python
# Análise da resposta para a pergunta 2 (Segunda parte)
# Plote um gráfico que mostre a distribuição de bons e maus pagadores por estado. Escolha o gráfico que achar mais conveniente


# Criando o gráfico de barras agrupado para mostrar a distribuição por estado
plt.figure(figsize=(15, 8))
sns.countplot(x='ESTADO_RESIDENCIAL', hue='classificacao', data=df)
plt.title('Distribuição de Bons e Maus Pagadores por Estado')
plt.xlabel('Estado de Residência')
plt.ylabel('Contagem')
plt.legend(title='Classificação', labels=['Bom Pagador (0)', 'Mau Pagador (1)'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Use boxplot para explicar como o tipo de residência interfere na classificação do pagador


# Criando um boxplot para explicar a relação entre o tipo de residência e a renda
# separada por classificação de pagador.
plt.figure(figsize=(15, 8))
sns.boxplot(x='TIPO_RESIDENCIA', y='RENDA_PESSOAL_MENSAL', hue='classificacao', data=df)
plt.title('Distribuição da Renda Pessoal Mensal por Tipo de Residência e Classificação')
plt.xlabel('Tipo de Residência')
plt.ylabel('Renda Pessoal Mensal')
plt.legend(title='Classificação', labels=['Bom Pagador (0)', 'Mau Pagador (1)'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Como as features renda pessoal mensal e outras rendas se relacionam com a variável meta? (obs: não é correlação estatística)

# Criando um boxplot para Renda Pessoal Mensal
plt.figure(figsize=(10, 6))
sns.boxplot(x='classificacao', y='RENDA_PESSOAL_MENSAL', data=df)
plt.title('Relação entre Renda Pessoal Mensal e Classificação do Pagador')
plt.xlabel('Classificação')
plt.ylabel('Renda Pessoal Mensal')
plt.xticks(ticks=[0, 1], labels=['Bom Pagador', 'Mau Pagador'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
